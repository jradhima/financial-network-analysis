```python
import os

count = 0

# traverse everything starting from folder 'sec-edgar-filings'
for pathnames, dirnames, filenames in os.walk('sec2017/sec-edgar-filings/'):
    # check if every file if it's a submission file
    for file in filenames:
        if file == 'full-submission.txtnew':
            
            count += 1
            
            filepath = pathnames + os.sep + file
            
            print(count, filepath)
```
