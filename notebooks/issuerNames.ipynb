```python
import pandas as pd
```

```python
df = pd.read_csv('filingsEnd2017.csv')
```

```python
df = df.drop(columns=['Unnamed: 0', 'value', 'amount', 'put_or_call', 'owner', 'cik', 'report_date'])
```

```python
df['issuer'] = df['cusip'].apply(lambda x: x[:6])
df
```

```python
df.groupby('issuer')['filed name'].count().sort_values(ascending=False)[:50]
```

```python
clean = df.drop_duplicates(subset='issuer')
```

```python
clean['label'] = clean['filed name'].apply(lambda x: ' '.join(x.split(' ')[:3]))
```

```python
clean.sample(20)
```

```python
clean.to_csv('cusipIssuer.csv')
```
